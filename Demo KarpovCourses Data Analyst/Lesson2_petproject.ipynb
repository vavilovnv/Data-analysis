```python
# https://stepik.org/lesson/375665/step/1?unit=363415

# Импортируйте библиотеку pandas как pd. Прочитайте датасет bookings.csv , находящийся по ссылке, с разделителем ;  и сохраните 
# в переменную bookings. Затем запишите первые 7 строк в переменную bookings_head

import pandas as pd
```

```python
bookings = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', sep=';')
bookings
```

```python
# bookings_head = bookings[:7]

bookings_head = bookings.head(n=7)
bookings_head
```

```python
# К какому типу относится большинство переменных? 

bookings.dtypes.value_counts()
```

```python
# Как вы уже могли заметить, при сборе датасета произошла небольшая ошибка, и названия колонок были записаны в неудобном для 
# дальнейшей работы формате. Исправьте ситуацию, приведя названия столбцов к нижнему регистру и заменив пробелы на знак нижнего 
# подчеркивания. Например: Is Canceled -> is_canceled

bookings = bookings.rename(columns = {column: column.replace(' ','_').lower() for column in bookings.columns})
bookings
```

```python
# Пользователи из каких стран совершили наибольшее число успешных бронирований? 
# Бронирование считается успешным, если в дальнейшем не было отменено (переменная is_canceled). В качестве ответа выберите 
# страны, входящие в топ-5.

(bookings.query('is_canceled == 0')
     .groupby('country')
     .agg({'is_canceled': 'count'})
     .sort_values('is_canceled', ascending = False)
     .head(5))
```

```python
# На сколько ночей (stays_total_nights)  в среднем бронируют отели типа City Hotel? Resort Hotel? Запишите полученные значения 
# в пропуски с точностью до 2 знаков после точки.

bookings.groupby('hotel').agg({'stays_total_nights': 'mean'}).round(2)
```

```python
# Иногда тип номера, полученного клиентом (assigned_room_type), отличается от изначально забронированного (reserved_room_type). 
# Такое может произойти, например, по причине овербукинга.
# Сколько подобных наблюдений встретилось в датасете?

len(bookings.query('reserved_room_type != assigned_room_type'))
```

```python
# Теперь проанализируйте даты запланированного прибытия (arrival_date_year).
# На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?

(bookings.groupby(["arrival_date_year", "arrival_date_month"])
     .agg({"is_canceled": "count"})
     .sort_values(["arrival_date_year", "is_canceled"], ascending = [True, False])
     .groupby("arrival_date_year")
     .head(1))
```

```python
# Сгруппируйте данные по годам, а затем проверьте, на какой месяц (arrival_date_month) бронирования отеля типа City Hotel 
# отменялись чаще всего в 2015? 2016? 2017? 

(bookings.query("hotel == 'City Hotel'")
     .groupby(["arrival_date_year", "arrival_date_month"])
     .agg({"is_canceled": "count"})
     .sort_values(["arrival_date_year", "is_canceled"], ascending = [True, False])
     .groupby("arrival_date_year")
     .head(1))
```

```python
# Посмотрите на числовые характеристики трёх колонок: adults, children и babies
# Какая из них имеет наибольшее среднее значение?

bookings.agg({"adults": "mean", "children": "mean", "babies": "mean"}).idxmax()
```

```python
# Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось
# наибольшим?
# City hotel – отель находится в городе
# Resort hotel – отель курортный
# В качестве ответа укажите наибольшее среднее total_kids, округлив до 2 знаков после точки.

bookings['total_kids'] = bookings.children + bookings.babies
```

```python
(bookings.groupby("hotel")
     .agg({"total_kids": "mean"})
     .round(2)
     .max())
```

```python
# Не все бронирования завершились успешно (is_canceled), поэтому можно посчитать, сколько клиентов было потеряно в процессе. 
# Churn rate (отток, коэффициент оттока) – это процент подписчиков (например, на push-уведомления от сайта), которые отписались 
# от канала коммуникации, отказались от услуг сервиса в течение определенного периода времени. Иными словами, представляет собой
# отношение количества ушедших пользователей к общему количеству пользователей, выраженное в процентах.
# Создайте переменную has_kids, которая принимает значение True, если клиент при бронировании указал хотя бы одного ребенка 
# (total_kids), в противном случае – False. Далее проверьте, среди каких пользователей показатель оттока выше. 
# В качестве ответа укажите наибольший %, округленный до 2 знаков после точки (то есть доля 0.24563 будет 24.56% и в ответ 
# пойдёт 24.56)

bookings['has_kids'] = bookings.total_kids > 0
```

```python
# Группа бронирований с детьми

round(bookings.query('has_kids and is_canceled == 1').shape[0] / bookings.query('has_kids').shape[0] * 100, 2)
```

```python
# Группа бронирований без детей

round(bookings.query('has_kids == False and is_canceled == 1').shape[0] / bookings.query('has_kids == False').shape[0] * 100, 2)
```
