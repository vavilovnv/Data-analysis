## Итерация по датафрейму (что нежелательно)

````python
# вывод названий колонок

for i in genres:
    print(i)
````

````python
# альтернативный вариант

for i in genres.columns:
    print(i)
````

````python
# вывод значений (без разбиений по ячейкам)

for i in genres.values:
    print(i)
````

````python
# вывод значений (с разбиением по ячейкам)

for row in genres.values:
    for value in row:
        print(value)
````

применение функций

````python
# создадим для примера функцию реверса строковых значений

def reversator(value):
    return value[::-1]
````

````python
# применим функцию к каждому их элементов датафрейма

for row in genres.values:
    for value in row:
        print(reversator(value))
````

````python
# альтернативный способ 

for i, row in genres.iterrows():
    print(row.map(reversator))
````

````python
# альтернативный способ

for i, col in genres.iteritems():
    print(col.map(reversator))
````

## Pandas-way (без итерирования по датафрейму)

````python
budget = df[['budget', 'duration']]
budget.head()
````

````python
# используем функцию applymap для увеличения всех значений на единицу (функция применима к каждому значению)

budget.applymap(lambda x: x+1)
````

````python
# используем функцию apply для расчета среднего по колонкам (axis=0 говорит о том, что функция применима к колонкам)

budget.apply(np.mean, axis=0)
````

````python
# используем функцию apply для расчета среднего по строкам (axis=0 говорит о том, что функция считает среднее по значениям
# ячеек в строках

budget.apply(np.mean, axis=1)
````

````python
# функция transform применяется к значениям датафрейма, но не применима в случае использования агрегирующих функций, таких как
# np.mean()

budget.transform(lambda x: x+1)
````

````python
# использование функций в apply

def mm(col):
    return np.mean(col) + 1
````

````python
budget.apply(mm)
````

````python
# более производительный вариант, без создания функции, используя встроенные функции pandas

budget.mean() + 1
````

````python
budget.max()
````

````python
# еще быстрее работать с numpy arrays, т.к. pandas более высокоуровневая либа по сравнению с numpy

budget.values
````

````python
# применим mean к array, но предварительно избавимся от NaN в значениях

np.mean(budget['budget'].dropna().values)
````

## Работа с timeseries данными

````python
stocks = pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/amzn_stock.csv')
stocks.head()
````

````python
stocks.index
````

````python
# считаем данные с timeseries
````

````python
stocks = pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/amzn_stock.csv', index_col='Date', parse_dates=True)
stocks.head()
````

````python
stocks.index
````

````python
# timeindex дает возможность отбора по индексу или его части

stocks['2010'].head() #отбирает все строки год в индексе которых равен 2010
````

````python
# отбор данных с февраля 2010 по март 2011

stocks['2010-02':'2011-03']
````

````python
# изменение масштаба выборки данных благодаря наличию timeseries в индексе
# например, разобьем данные с интервалом в 2 часа

stocks.resample('2h').asfreq()
````

````python
# свернем данные с интервалом в 1 неделю и агрегируем, получив средние значения понедельно

stocks.resample('1w').mean().head()
````

## Работа с "окнами-выборками" данных

````python
# обработка данных "окном", т.е. операция расчета применяется к текущей строке и n-1 количеству предыдущих строк
# например, применим расчет среднего к текущей строке и двум предыдущим

stocks.rolling(3).mean().head(7)
````

````python
# уберем из вывода строки без данных

stocks.rolling(3, min_periods=1).mean().head(7)
````

````python
# обработка данных "разрастающимся окном", т.е. операция расчета применяется, если не указаны параметры как-бы разрастаясь,
# от первой ко второй строке, от первой к третьей, от первой к четвертой и т.д. Параметр ограничивает минимальное количество 
# после которых начнется расчет

stocks.expanding(3).mean().head(7)
````

````python
# экспоненциально взвешенное окно, т.е. при расчете используются не одинаковые веса для каждой строки, а в зависимости от 
# текущей строки. У текущей строки максимальный вес, у предыдущей поменьше и т.д.

stocks.ewm(3).mean().head(7)
````
